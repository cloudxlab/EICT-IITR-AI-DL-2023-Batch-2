```python
import pandas as pd
pointsdf = pd.read_csv("/content/sample_data/points.csv");

import numpy as np
points = np.array(pointsdf)

xs = points[:, 0]
ys  = points[:, 1]

import matplotlib.pyplot as plt
plt.scatter(xs, ys)
```

**1.** **K-means clustering algorithm download from the SK learn.**

**2.** **A K-Means clustering model is created with the specified number of clusters, which is set to 3. This means the algorithm will try to cluster the data points into three distinct groups.**


**3.** **Predict the cluster labels for the new data points new_points using the previously trained KMeans model.**


**4.** **The predicted cluster labels for the new data points are printed to the console.**

```python
from sklearn.cluster import KMeans
model = KMeans( n_clusters=3)
t=model.fit(points)
clusters = model.fit_predict(points)
clusters
new_points = [[4,3],[4,1],[-5,1]]
j = model.predict(new_points)
print(j)
```

# "elbow method" for determining an appropriate number of clusters (k) in a KMeans clustering algorithm.

#"elbow curve" that helps you visually determine an appropriate number of clusters for your data.
#The idea is to look for the "elbow point" on the curve where the inertia starts to decrease at a slower rate. This point often corresponds to a suitable number of clusters, as adding more clusters beyond this point may not significantly improve the clustering quality.

```python
ks = range(1, 10)
inertias = [] # An empty list named inertias is created to store the inertia values for each value of k.

for k in ks:
    # Create a KMeans instance with k clusters: model
    model = KMeans(n_clusters=k)

    # Fit model to samples
    model.fit(samples)

    # Append the inertia to the list of inertias
    inertias.append(model.inertia_)

# Plot ks vs inertias
plt.plot(ks, inertias, '-o')
plt.xlabel('number of clusters, k')
plt.ylabel('inertia')
plt.xticks(ks)
plt.show()
```

## Using Cross Tab - in this method creating a cross-tabulation (crosstab) of the cluster labels and the original varieties or categories of the data.
## This method helps you understand how well the KMeans clustering algorithm has grouped data points into clusters and how these clusters relate to the known categories or varieties of the data.

```python

samples = data[:, 0:7]
varieties = data[:, 7]

# Create a KMeans model with 3 clusters: model
model = KMeans(n_clusters=3)

# Use fit_predict to fit model and obtain cluster labels: labels
labels = model.fit_predict(samples)

# Create a DataFrame with labels and varieties as columns: df
df = pd.DataFrame({'labels': labels, 'varieties': varieties})

# Create crosstab: ct
ct = pd.crosstab(df['labels'], df['varieties'])

# Display ct
print(ct)
```
